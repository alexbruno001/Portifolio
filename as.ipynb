```python
import pandas as pd
```

```python
#carregando dados
df= pd.read_csv("./2004-2021.tsv", sep= "\t")
df.head()
```

```python
#Alterando o tipo dos dados
df["DATA FINAL"] = pd.to_datetime(df["DATA FINAL"])
df["DATA INICIAL"]= pd.to_datetime(df["DATA INICIAL"])
df.info()
```

```python
#Alterando os atributos para numéricos
for atributo in['MARGEM MÉDIA REVENDA', 'PREÇO MÉDIO DISTRIBUIÇÃO', 'DESVIO PADRÃO DISTRIBUIÇÃO', 'PREÇO MÍNIMO DISTRIBUIÇÃO', 'PREÇO MÁXIMO DISTRIBUIÇÃO', 'COEF DE VARIAÇÃO DISTRIBUIÇÃO'] :
    df[atributo] = pd.to_numeric(df[atributo], errors= "coerce")
    df.info()
```

```python
#consultando dados sem salvar nenhuma variavel
df.query('REGIÃO in["CENTRO OESTE", "SUL", "NORDESTE"]')
```

```python
#filtrando por regiões
selec_1 = df["REGIÃO"] == "SUL"
selec_2 = df["REGIÃO"] == "CENTRO OESTE"

df_selec_regiao = selec_1 | selec_2
df[df_selec_regiao]
```

```python
#verificando valores nulos
df["PREÇO MÁXIMO DISTRIBUIÇÃO"].isnull().sum()
```

```python
#removendo dados nulos na tabela teste
df_teste= df.dropna()
df_teste
```

```python
#retorna os valores nulos
df_nulos = df[df.isnull().any(axis=1)]
df_nulos
```

```python
#Retorna a linhas sem os valores nulos
df_teste_1 = df.dropna(axis=0)
df_teste_1
```

```python
#Retorna as colunas sem os valores nulos
df_teste_2 = df.dropna(axis=1)
df_teste_2
```

```python
#media dos valores
df_teste_4 = df.fillna(df.mean(numeric_only=True))
df_teste_4.mean(numeric_only=True)
```

```python
#localizando valores nulos
df_nulos = df["PREÇO MÉDIO DISTRIBUIÇÃO"].isnull()
df[df_nulos]
```
